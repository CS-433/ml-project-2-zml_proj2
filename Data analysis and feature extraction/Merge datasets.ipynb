```python
import os
from tqdm.notebook import tqdm
import gzip
import pandas as pd
import seaborn as sns
import timeit
import json
import matplotlib.pyplot as plt
import numpy as np
```

# Load Data

```python
DATA_FOLDER = "./"
DATA_SRC1 = DATA_FOLDER + "tweets-1.json"
DATA_SRC2 = DATA_FOLDER + "tweets-2.json"
DATA_SRC3 = DATA_FOLDER + "tweets-3.json"
DATA_SRC4 = DATA_FOLDER + "tweets-4.json"
```

## Tweets 1/4

```python
data = pd.read_json(DATA_SRC1, lines=True, nrows=100)
data.head()
```

```python
data1 = {}
for column in data:
    data1[column] = list(data[column].at[0].values())

data1 = pd.DataFrame(data1)
data1.head()
```

## Tweets 2/4

```python
data = pd.read_json(DATA_SRC2, lines=True, nrows=100)
data.head()
```

```python
data2 = {}
for column in data:
    data2[column] = list(data[column].at[0].values())

data2 = pd.DataFrame(data2)
data2.head()
```

## Tweets 3/4

```python
data = pd.read_json(DATA_SRC3, lines=True, nrows=100)
data.head()
```

```python
data3 = {}
for column in data:
    data3[column] = list(data[column].at[0].values())

data3 = pd.DataFrame(data3)
data3.head()
```

## Tweets 4/4

```python
data = pd.read_json(DATA_SRC4, lines=True, nrows=100)
data.head()
```

```python
data4 = {}
for column in data:
    data4[column] = list(data[column].at[0].values())

data4 = pd.DataFrame(data4)
data4.head()
```

## Merge data

```python
data = pd.concat([data1, data2, data3, data4], ignore_index=True)
data.head()
```

```python
data.info()
```

## Save data

```python
data.to_json("tweets.json")
```
